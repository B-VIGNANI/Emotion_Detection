```python
from tkinter import *
from tkinter.messagebox import showinfo
from gtts import gTTS
import speech_recognition as sr
import os
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Global flag for dark mode
is_dark_mode = False

# Theme colors
LIGHT_THEME = {'bg': 'white', 'fg': 'black'}
DARK_THEME = {'bg': '#2b2b2b', 'fg': 'white'}

# Main Window
mainwindow = Tk()
mainwindow.title('Text Emotion Recognition')
mainwindow.geometry('500x500')
mainwindow.configure(bg=LIGHT_THEME['bg'])

# Center heading
Label(
    mainwindow,
    text='Text Emotion Recognition',
    font=('Times New Roman', 20, 'bold'),
    bg=LIGHT_THEME['bg'],
    fg=LIGHT_THEME['fg']
).place(relx=0.5, y=20, anchor='center')


def toggle_dark_mode():
    global is_dark_mode
    is_dark_mode = not is_dark_mode
    theme = DARK_THEME if is_dark_mode else LIGHT_THEME
    mainwindow.configure(bg=theme['bg'])
    for widget in mainwindow.winfo_children():
        try:
            widget.configure(bg=theme['bg'], fg=theme['fg'])
        except:
            pass


def say(text1):
    language = 'en'
    speech = gTTS(text=text1, lang=language, slow=False)
    speech.save("text.mp3")
    os.system("start text.mp3")


def recordvoice():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        audio = r.listen(source)
        try:
            text = r.recognize_google(audio, language="en-IN")
        except:
            text = "Sorry, could not recognize."
        return text


def clearAll():
    negativeField.delete(0, END)
    neutralField.delete(0, END)
    positiveField.delete(0, END)
    overallField.delete(0, END)
    textArea.delete(1.0, END)


def detect_sentiment():
    sentence = textArea.get("1.0", "end")
    sid_obj = SentimentIntensityAnalyzer()
    sentiment_dict = sid_obj.polarity_scores(sentence)

    negativeField.insert(10, str(sentiment_dict['neg']*100) + "% Negative")
    neutralField.insert(10, str(sentiment_dict['neu']*100) + "% Neutral")
    positiveField.insert(10, str(sentiment_dict['pos']*100) + "% Positive")

    if sentiment_dict['compound'] >= 0.05:
        string = "Positive"
    elif sentiment_dict['compound'] <= -0.05:
        string = "Negative"
    else:
        string = "Neutral"

    overallField.insert(10, string)


def Sentiment_Analysis():
    global textArea, negativeField, neutralField, positiveField, overallField

    gui = Toplevel(mainwindow)
    gui.title("Emotion Recognition")
    gui.geometry("300x500")
    gui.config(bg=LIGHT_THEME['bg'])

    Button(gui, text='â† Back', bg='gray', command=gui.destroy).place(x=5, y=5)

    Label(gui, text="Enter Your Sentence", bg='white').pack(pady=10)
    textArea = Text(gui, height=5, width=30, font="lucida 13")
    textArea.pack()

    Button(gui, text="Check Emotion", bg="#00eb17", command=detect_sentiment).pack(pady=10)

    Label(gui, text="Negative:", bg='white').pack()
    negativeField = Entry(gui)
    negativeField.pack()

    Label(gui, text="Neutral:", bg='white').pack()
    neutralField = Entry(gui)
    neutralField.pack()

    Label(gui, text="Positive:", bg='white').pack()
    positiveField = Entry(gui)
    positiveField.pack()

    Label(gui, text="Overall:", bg='white').pack()
    overallField = Entry(gui)
    overallField.pack()

    Button(gui, text="Clear", bg="white", command=clearAll).pack(pady=5)
    Button(gui, text="Exit", bg="red", command=gui.destroy).pack()


def TextToSpeech():
    texttospeechwindow = Toplevel(mainwindow)
    texttospeechwindow.title('Text-to-Speech Converter')
    texttospeechwindow.geometry("400x300")
    texttospeechwindow.configure(bg='white')

    Button(texttospeechwindow, text='â† Back', bg='gray', command=texttospeechwindow.destroy).place(x=5, y=5)

    Label(texttospeechwindow, text='Text-to-Speech Converter', font=("Comic Sans MS", 15), bg='white').pack(pady=10)
    text = Text(texttospeechwindow, height=5, width=30, font=12)
    text.pack()

    Button(texttospeechwindow, text='Listen', bg='coral', command=lambda: say(str(text.get(1.0, END)))).pack(pady=10)


def SpeechToText():
    speechtotextwindow = Toplevel(mainwindow)
    speechtotextwindow.title('Speech-to-Text Converter')
    speechtotextwindow.geometry("400x300")
    speechtotextwindow.configure(bg='white')

    Button(speechtotextwindow, text='â† Back', bg='gray', command=speechtotextwindow.destroy).place(x=5, y=5)

    Label(speechtotextwindow, text='Speech-to-Text Converter', font=("Comic Sans MS", 15), bg='white').pack(pady=10)

    text = Text(speechtotextwindow, font=12, height=3, width=30)
    text.pack(pady=10)

    Button(speechtotextwindow, text='Record', bg='#db327e', command=lambda: text.insert(END, recordvoice())).pack()


# Menu Bar
menubar = Menu(mainwindow)
mainwindow.config(menu=menubar)

def show_manual():
    showinfo("User Manual", "1. Click 'Emotion Recognition Station' to analyze text emotion.\n"
             "2. Use 'Text-To-Speech' to listen to typed text.\n"
             "3. Use 'Speech-To-Text' to convert your voice into text.\n"
             "4. Use the 'Dark Mode' toggle to change themes.\n"
             "5. Exit any window with the 'â† Back' button.")

def show_about():
    showinfo("About the App", "Text Emotion Recognition App\n"
             "Developed using Python (Tkinter)\n"
             "Features: Text Emotion Detection, TTS, STT, Dark Mode\n"
             "Sentiment Analysis using VADER")

helpmenu = Menu(menubar, tearoff=0)
helpmenu.add_command(label="User Manual", command=show_manual)
helpmenu.add_command(label="About", command=show_about)
helpmenu.add_separator()
helpmenu.add_command(label="Exit", command=mainwindow.quit)
menubar.add_cascade(label="Help", menu=helpmenu)

# Buttons
Button(mainwindow, text='Emotion Recognition Station', font=('Times New Roman', 14), bg='#20f200',
       command=Sentiment_Analysis).place(x=100, y=100)

Button(mainwindow, text='Text-To-Speech Conversion', font=('Times New Roman', 14), bg='#fc9803',
       command=TextToSpeech).place(x=100, y=180)

Button(mainwindow, text='Speech-To-Text Conversion', font=('Times New Roman', 14), bg='#9329f0',
       command=SpeechToText).place(x=100, y=260)

Button(mainwindow, text='Toggle Dark Mode', font=('Times New Roman', 12), bg='gray',
       command=toggle_dark_mode).place(x=160, y=340)

mainwindow.mainloop()
```

```python
from tkinter import *
from tkinter import filedialog, messagebox
from tkinter import ttk
from gtts import gTTS
import speech_recognition as sr
import os
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from matplotlib import pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from fpdf import FPDF
from googletrans import Translator
import json

# Initialize translator
translator = Translator()

# Save and load paths
SAVE_FILE = "emotion_notes.json"

# Main Window
mainwindow = Tk()
mainwindow.title('ðŸ§  Text Emotion Recognition')
mainwindow.geometry('700x700')
mainwindow.configure(bg='white')

# Dark Mode State
is_dark_mode = False

# Toggle Dark Mode
def toggle_dark_mode():
    global is_dark_mode
    is_dark_mode = not is_dark_mode
    bg_color = '#333333' if is_dark_mode else 'white'
    fg_color = 'white' if is_dark_mode else 'black'
    mainwindow.configure(bg=bg_color)
    heading_label.configure(bg=bg_color, fg=fg_color)

# Back to main menu
def go_back(win):
    win.destroy()
    mainwindow.deiconify()

# Save Notes
def save_notes(text, emotion):
    data = {"text": text, "emotion": emotion}
    with open(SAVE_FILE, 'w') as f:
        json.dump(data, f)
    messagebox.showinfo("âœ… Saved", "Notes saved successfully!")

# Load Notes
def load_notes():
    if os.path.exists(SAVE_FILE):
        with open(SAVE_FILE, 'r') as f:
            data = json.load(f)
        return data.get("text", ""), data.get("emotion", "")
    return "", ""

# Export to PDF
def export_to_pdf(text, sentiment_dict):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=12)
    pdf.cell(200, 10, txt="ðŸ§  Sentiment Analysis Report", ln=True, align='C')
    pdf.ln(10)
    pdf.multi_cell(0, 10, txt=f"Text:\n{text}\n")
    pdf.ln(5)
    for key, val in sentiment_dict.items():
        pdf.cell(0, 10, f"{key.capitalize()} Score: {val}", ln=True)
    pdf.output("Sentiment_Report.pdf")
    messagebox.showinfo("ðŸ“¤ Exported", "PDF Exported Successfully!")

# Plot Emotion Chart
def plot_chart(frame, scores):
    fig = plt.Figure(figsize=(3.5, 2.5), dpi=100)
    ax = fig.add_subplot(111)
    categories = list(scores.keys())[:-1]
    values = [scores[k] for k in categories]
    ax.bar(categories, values, color=['red', 'gray', 'green'])
    ax.set_title('ðŸ“Š Emotion Score Chart')
    canvas = FigureCanvasTkAgg(fig, master=frame)
    canvas.draw()
    canvas.get_tk_widget().pack()

# Text to Speech
def text_to_speech(text):
    tts = gTTS(text=text, lang='en')
    tts.save("speech.mp3")
    os.system("start speech.mp3")

# Speech to Text
def speech_to_text():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        messagebox.showinfo("ðŸŽ¤ Speak Now", "Please speak...")
        audio = recognizer.listen(source)
        try:
            text = recognizer.recognize_google(audio)
            messagebox.showinfo("ðŸ—£ï¸ Detected Text", text)
        except sr.UnknownValueError:
            messagebox.showerror("âŒ Error", "Could not understand audio")
        except sr.RequestError:
            messagebox.showerror("âŒ Error", "Could not request results")

# Translate Text
def translate_text(text):
    translated = translator.translate(text, dest='ta')
    messagebox.showinfo("ðŸŒ Translated Text", translated.text)

# About and User Manual
def show_about():
    messagebox.showinfo("ðŸ“˜ About", "This app analyzes emotional tone in text using VADER sentiment analysis. It supports visualization, saving, exporting, translation, and speech features.")

def show_manual():
    messagebox.showinfo("ðŸ“– User Manual", "1. Type your text\n2. Click 'Analyze Emotion'\n3. View chart and results\n4. Save, export, translate or use speech tools from the menu.")

# Emotion Recognition Window
def Sentiment_Analysis():
    mainwindow.withdraw()
    gui = Toplevel()
    gui.title("ðŸ§  Emotion Recognition Station")
    gui.geometry("550x750")
    gui.configure(bg='white')

    Label(gui, text="ðŸ§  Emotion Recognition Station", font=("Arial", 16, "bold"), bg='white').pack(pady=10)
    Button(gui, text='â¬…ï¸ Back', bg='lightgray', command=lambda: go_back(gui)).place(x=5, y=5)

    Label(gui, text="âœï¸ Enter your sentence:", bg='white').pack(pady=5)
    textArea = Text(gui, height=5, width=50, font="Arial 12")
    textArea.pack(pady=5)

    result_frame = Frame(gui, bg='white')
    result_frame.pack(pady=5)

    def detect_sentiment():
        sentence = textArea.get("1.0", "end").strip()
        if not sentence:
            return messagebox.showwarning("âš ï¸ Warning", "Please enter some text")
        sid_obj = SentimentIntensityAnalyzer()
        sentiment_dict = sid_obj.polarity_scores(sentence)

        for widget in result_frame.winfo_children():
            widget.destroy()

        Label(result_frame, text=f"ðŸ“‰ Negative: {sentiment_dict['neg']*100:.2f}%", bg='white').pack()
        Label(result_frame, text=f"ðŸ˜ Neutral: {sentiment_dict['neu']*100:.2f}%", bg='white').pack()
        Label(result_frame, text=f"ðŸ˜Š Positive: {sentiment_dict['pos']*100:.2f}%", bg='white').pack()

        compound = sentiment_dict['compound']
        if compound >= 0.05:
            emotion = "ðŸ˜Š Positive"
        elif compound <= -0.05:
            emotion = "â˜¹ï¸ Negative"
        else:
            emotion = "ðŸ˜ Neutral"

        Label(result_frame, text=f"ðŸ§¾ Overall Sentiment: {emotion}", bg='white', font=("Arial", 12, "bold")).pack(pady=5)
        plot_chart(result_frame, sentiment_dict)

        Button(result_frame, text="ðŸ’¾ Save Note", command=lambda: save_notes(sentence, emotion)).pack(pady=2)
        Button(result_frame, text="ðŸ“¤ Export to PDF", command=lambda: export_to_pdf(sentence, sentiment_dict)).pack(pady=2)
        Button(result_frame, text="ðŸŒ Translate to Tamil", command=lambda: translate_text(sentence)).pack(pady=2)
        Button(result_frame, text="ðŸ”Š Text to Speech", command=lambda: text_to_speech(sentence)).pack(pady=2)

    Button(gui, text="ðŸ§  Analyze Emotion", bg="#00eb17", command=detect_sentiment).pack(pady=10)

    def load_btn():
        txt, emo = load_notes()
        textArea.delete(1.0, END)
        textArea.insert(END, txt)
        messagebox.showinfo("ðŸ“¥ Loaded", f"Previous Emotion: {emo}")

    Button(gui, text="ðŸ“¥ Load Saved Note", command=load_btn).pack(pady=5)

    Label(gui, text="ðŸ”Š Text-to-Speech Conversion", font=("Arial", 12, "bold"), bg='white').pack(pady=10)
    tts_entry = Entry(gui, width=40)
    tts_entry.pack()
    Button(gui, text="ðŸŽ§ Listen", command=lambda: text_to_speech(tts_entry.get())).pack(pady=2)

    Label(gui, text="ðŸŽ¤ Speech-to-Text Conversion", font=("Arial", 12, "bold"), bg='white').pack(pady=10)
    Button(gui, text="ðŸŽ™ï¸ Record & Transcribe", command=speech_to_text).pack(pady=2)

# UI Heading
heading_label = Label(mainwindow, text='ðŸ§  Text Emotion Recognition', font=('Arial', 20, 'bold'), bg='white')
heading_label.pack(pady=30)

# Main Buttons
Button(mainwindow, text='ðŸ” Emotion Recognition Station', font=('Arial', 14), bg='#20f200', command=Sentiment_Analysis).pack(pady=10)

# Menu Bar
menubar = Menu(mainwindow)
filemenu = Menu(menubar, tearoff=0)
filemenu.add_command(label="ðŸ“– User Manual", command=show_manual)
filemenu.add_command(label="ðŸ“˜ About App", command=show_about)
filemenu.add_command(label="ðŸ”Š Text to Speech", command=lambda: text_to_speech("Hello! Enter your own text to try it out."))
filemenu.add_command(label="ðŸŽ¤ Speech to Text", command=speech_to_text)
filemenu.add_separator()
filemenu.add_command(label="ðŸšª Exit", command=mainwindow.quit)
menubar.add_cascade(label="â˜° Menu", menu=filemenu)
menubar.add_command(label="ðŸŒ™ Toggle Dark Mode", command=toggle_dark_mode)
mainwindow.config(menu=menubar)

mainwindow.mainloop()
```
